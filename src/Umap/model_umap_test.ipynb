```python
import mapper

# paths
data_path: str       = "data"
model_path: str      = "model"
parameters_path: str = "params" 

data, in_dim, out_dim = mapper.fetch_data(path=data_path)

model = mapper.fetch_model(
		model_path=model_path, 
		parameters_path=parameters_path, 
		in_dim=in_dim,
		out_dim=out_dim
	)

embedding, labels = mapper.make_umap(
		model=model, 
		plans=data, 
		sample_amnt=1, 
		label_func=mapper.make_validity_labels
	)

mapper.plot(embedding=embedding, labels=labels)
```
